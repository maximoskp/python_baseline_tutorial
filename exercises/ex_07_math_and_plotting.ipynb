### Exercise 1

Numpy has a function that creates an array of random integers within a given range.

a) Create an array that simulates 1000 6-edge dice rolls.

b) Find the number of times the next dice roll gave an equal outcome to its next roll.

c) Find the maximum number of equal rolls in a row.

d) What happens to the numbers of b) and c) if you do 1000 times the dice rolls you did in a), i.e., 1000000 (one million) dice rolls?

Hint: in b) and c) you can use the numpy functions *diff* and *unique*.

Continuing from the above exercise (1), create a bar plot that shows the percentage of times that the next dice roll outcome differed by -5, -4, ..., 0, 1, ..., 4, 5 from its next. You can use the one million-times simulation for this plot.
