### Exercise 1

Create a class (e.g. "Book") for helping a bookstore keep count of the books in its inventory. Each book object in the class should be initialized with the following elements:
1. a title,
2. an author and an ISBN and
3. a count of how many copies there are of it in the inventory of the bookstore.

There should be a function is called when a copy of the book is purchased. This function should subtract one from the copies of the book in the inventory and it should print a warning message if the number of copies is smaller than or equal to 3, e.g., the message could be "There are 3 copies of the book in the inventory".

### Exercise 2

In the above exercise, add three static variables that count:

1. how many books there are in total in the inventory,
2. what is the number of copies and the titles of the book with the most and the least available copies in the inventory and
3. create a method that prints those statistics (i.e., the number of total copies of all books and the number and titles of the books with the most and least available copies in the bookstore).

Be caution regarding point 2: whenever a book is added or whenever a copy of a book is purchased, there should be proper functions that recalculate the statistics.
