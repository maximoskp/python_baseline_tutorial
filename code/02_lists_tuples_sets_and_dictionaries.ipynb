```python
# lists
x = [1,3,5,6.4, 'my name is Alice', [2,4,6]]
# length of list
print(x)
print(len(x))
# retrieving list elements - either one or multiple elements
print( x[0], x[2], x[4] )
print(x[5][1])
# find index of item in list
print( x.index(5) )
print( x.index( 'my name is Alice' ) )
# print(x.index(100)) # ValueError: 100 is not in list
# assign element at index
x[2] = 'New element'
print(x)
```

```python
# "summing" lists in python is concatenation (as in strings)
x = [2,4,6]
y = [3,5,7]
z = x + y
print(z)
# "multiplication" with integer is multiple concatenations
w = 3*x
print(w)
# u = x*y # TypeError: can't multiply sequence by non-int of type 'list'
```

```python
# append item in list
x = [1,4,7]
print(x)
x.append(1.4)
print(x)
y = [3,6,8]
# x.append(y) # appends one new component, the entire list

# extend list with another list (vs append another list to a list)
# 1
x.extend( y )
print(x)
# we can achieve the same result using '+'
# 2
# x = x + y
# print(x)
# 3
# x += y
# print(x)

# pop: remove element from position
x.pop(3)
print(x)
# remove: remove element by name (first occurance)
x.remove(4)
print(x)
```

```python
# range: e.g., r = range(10) # 0 to 9 - 10 not included
r = range(2,10,3)
# list form of range
x = list( r )
print(x)
# short for tutorial
for x in range(2,10,3):
    print(x)
```

```python
# tuples
# immutable lists
t = (1,3,5,4,3,6,7, 'string variable')
print(t)
# index works for tuples
print( t.index(3) )
# len works for tuples
print(len(t))
# count how many occurances of element in tuple
print(t.count(3))
```

```python
# append does not work
# t.append(6) # AttributeError: 'tuple' object has no attribute 'append'
# cannot change item in index
# t[0] = 10 # TypeError: 'tuple' object does not support item assignment
```

```python
# sets
# https://www.w3schools.com/python/python_ref_set.asp
# no dublicates are allowed
s = {1,3,5,3}
print(s)
```

```python
# s.append(6) # AttributeError: 'set' object has no attribute 'append'
# add element to set - cannot append
# subscripts do not work on sets
# print(s[2]) # TypeError: 'set' object is not subscriptable
```

```python
# special set operation
s1 = {1,2,3,4,5,6}
s2 = {4,5,6,7,8,9}
# union
s = s1.union(s2)
print(s)
# intersection
s = s1.intersection(s2)
print(s)
# difference (non simmetric)
s = s1.difference(s2)
print(s)
# shorthands
# difference "-", subset "<=", symmetric difference "^"
s = s2 - s1
print(s)
print( s1 <= s2 )
print(s <= s2)
s = s1^s2
print(s)
```

```python
# interchanging between types
print(s)
x = list(s)
print(x)
t = tuple(s)
print(t)
y = [1,2,3,4,3,4,5]
print(set(y))
```

```python
# max, min, sort
x = [6,3,7,9,2,12]
print(max(x))
print(min(x))
x.sort()
print(x)
```

```python
# dictionary
# initialization with some values
d = {
    'name': 'Alice',
    'age': 20,
    'height': 1.7
}
print(d)
# empty initialization using {}
g = {}
print(g)
# keys and values
k = list( d.keys() )
print(k)
v = list( d.values() )
print(v)
```

```python
# retrieve value with key
print( d['name'] )
# add information by using [] with new key
d['fav_foods'] = ['pizza', 'spaghetti']
print(d)
g['Brand'] = 'Ford'
g['Model'] = 'Fiesta'
print(g)
```

```python
# delete information with del
del( d['height'] )
print(d)
```
