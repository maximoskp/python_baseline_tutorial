```python
# functions with no input and no output
# e.g., simple greeting
def simple_greeting():
    print('Hi!')
    print('This is a simple greeting')
    print('yup')
```

```python
simple_greeting()
```

```python
# functions with input and no output - the name of the function does not matter
# e.g., simple greeting with given name
def simple_greeting_with_name(x):
    print('Hi', x, '!')
    print('This is a simple greeting')
    print('yup')
```

```python
simple_greeting_with_name('Clara')
```

```python
# functions with no input but with output
# e.g., function that creates the Fibonacci numbers up to 100
def fibonacci():
    x = []
    a, b = 0, 1
    while a < 100:
        x.append(a)
        a, b = b, a+b
    return x
```

```python
y = fibonacci()
print(y)
```

```python
# functions with input and output
# e.g., function that creates the Fibonacci numbers up to a given number
def fibonacci_up_to_number(n):
    x = []
    a, b = 0, 1
    while a < n:
        x.append(a)
        a, b = b, a+b
    return x
```

```python
y = fibonacci_up_to_number(10000)
print(y)
```

```python
# multiple inputs and multiple outputs
# e.g., a function that receives two numbers and outputs the basic arithmetic
# operations
def basic_arithmetic_ops(x, y):
    summation = x + y
    subtraction = x - y
    multiplication = x * y
    division = x / y
    return summation, subtraction, multiplication, division
```

```python
sm, sb, ml, dv = basic_arithmetic_ops( 5,6 )
```

```python
# get outputs in a single tuple
y = basic_arithmetic_ops( 5,6 )
```

```python
# ignoring specific output in output position using "_"
a1, _, a2, _ = basic_arithmetic_ops( 5,6 )
```
