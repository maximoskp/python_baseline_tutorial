```python
# variables and print
# integer variables
x = 5
y = 4
# basic operations with integers (add, subtract, multiply, divide, power)
z = x + y
print(x)
print(y)
print( x-y )
print( x*y )
print( x/y )
print( x**y )
```

```python
# multiple assignments
# semicolon
x = 7 ; y = 9
# comma
x , y = 5 , 8
```

```python
# float variables
x = 4.5
y = 5
# basic operations as in integers
# basic operations with integers, leads to float
z = x + y
```

```python
# string variables
s1 = 'My name is '
s2 = 'Alice.'
# '+' in strings means concatenation
s = s1 + s2
print(s)
# '*' between string and integer means times of repetition
s = 3*s1
print(s)
# other operations have no meaning...
# s = s1*s2 # TypeError: can't multiply sequence by non-int of type 'str'
```

```python
# Boolean variables
b = False
a = True
# logical operations (and, or, not)
print( b and a )
print( b or a )
print( b )
print( not b )
# numerical operations (True becomes 1, False becomes 0)
print( b + 4 )
print( a + 4 )
```

```python
# identify type using the "type" function
print( type( a ) )
print( type( x ) )
print( type( y ) )
print( type( s ) )
```

```python
# change type of variable
# str, int, float, bool
y1 = float(y)
print(y1)
print(x)
x1 = int(x)
print(x1)
x_string = str(x)
print(type(x_string))
# isdigit function: check if string is digit
# 'My name is Bob.'.isdigit() # False
# '4'.isdigit() # True
```

```python
# special string functions
s = 'This - string - is - separated - by - hyphens.'
# split
splt = s.split(' -')
print(splt)
# replace
rplc = s.replace('- ', '*=')
print(rplc)
# lower and upper
print( s.lower() )
print( s.upper() )
```
