```python
# logical expressions in python
x = 3
print( x <= 3 or x > 10  )
# >. >=, ...
# double '=', i.e. '==', means check, single '=' means assignment
print( not x == 4 )
# check if item in list or not in list
c = [3,5,7]
print(c)
print( x in c )
print( (not (-1 in c)) and (x in c) )
```

```python
# if condition
# condition = x > 10
if x < 10:
    d = ' Wow!'
    print("It's True!" + d)
print('After the if statement')
```

```python
# if - else
t = 25
if t < 10:
    print('It is cold in here!')
else:
    print('The temperature is fine.')
```

```python
# if - elif - else
t = 35
if t < 10:
    print('It is cold.')
elif t < 24:
    print('It is normal.')
elif t < 30:
     print('It is starting to get hot...')
else:
    print('It is hot!')
```

```python
# for loops in python - iterating through lists
a = [ 5, 7, 'my name is Alice', [2,4,6] ]
for x in a:
    print(x)
```

```python
# for loops in range
for i in range(2,10,4):
    print(i)
```

```python
a = [34, 45, 56, 67, 78]
for i in range(5):
    if a[i] > 60:
        break
    print( 'a[' + str(i) + '] = ' + str(a[i]) )
```

```python
# iterate though a list using indexes with range and len
a = [34, 36, 45, 56, 67, 78, 100]
for i in range( len(a) ):
    # if a[i] > 60:
    #     break
    print( 'a[' + str(i) + '] = ' + str(a[i]) )
```

```python
# iterate though a list using indexes with enumerate
a = [34, 36, 45, 56, 67, 78, 100]
for i, x in enumerate(a):
    print( 'a[' + str(i) + '] = ' + str(x) )
```
